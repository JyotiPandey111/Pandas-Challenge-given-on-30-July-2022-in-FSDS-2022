## 5.Pre-processing of datasets
### 5.1 Pre-processing of Agent Performance Dataset
 

```python
agent_perf.dtypes
```

```python
Date = pd.to_datetime(agent_perf['Date'],dayfirst=True)
agent_perf['Date'] = Date

agent_perf['Week'] = agent_perf['Date'].dt.week

l = agent_perf['Average Response Time'].str.split(":")
agent_perf['Average Response Time(Seconds)'] = (l.str[0].astype(int) * 3600 + l.str[1].astype(int) * 60 + l.str[2].astype(int))
agent_perf['Average Response Time'] = pd.to_timedelta(agent_perf['Average Response Time'])


l = agent_perf['Average Resolution Time'].str.split(":")
agent_perf['Average Resolution Time(Seconds)'] = (l.str[0].astype(int) * 3600 + l.str[1].astype(int) * 60 + l.str[2].astype(int))
agent_perf['Average Resolution Time'] = pd.to_timedelta(agent_perf['Average Resolution Time'])

agent_perf.dtypes
```

```python
agent_perf.head()
```

### 5.2 Pre-processing of Agent Login Report Dataset

```python
agent_login.dtypes
```

```python
D = pd.to_datetime(agent_login['Date'],dayfirst=True)
agent_login['Date'] = D

agent_login['Week'] = agent_login['Date'].dt.week

L = pd.to_datetime(agent_login['Login Time'])
agent_login['Login Time'] = L

q = pd.to_datetime(agent_login['Logout Time'])
agent_login['Logout Time'] = q

l = agent_login['Duration'].str.split(":")
agent_login['Duration(hours)'] = (l.str[0].astype(int) * 3600 + l.str[1].astype(int) * 60 + l.str[2].astype(int)) / 3600
agent_login['Duration'] = pd.to_timedelta(agent_login['Duration'])

agent_login.dtypes
```

### 1 .Find out there average rating on weekly basis keep this in a mind that they take two days of leave

```python
agent_perf
```

```python
# considering chats where total number of the chat is greater than 0.
agent_perf_chats = agent_perf[agent_perf['Total Chats'] > 0] 

# grouping the chats on weeks and Agent Name on average ratings
df1 =  agent_perf_chats.groupby(['Week','Agent Name'])['Average Rating'].mean()
weekly_report = df1.to_frame()
weekly_report.head(50)
```

## 12. Percentage of chat on which they have received a feedback.
- Percentage of chat on which agents have received a feedback.
- Agent wise chat feedback percentage

```python
agent_perf_chats.head(2)
```

```python
# creating a dataframe df12 with total number of chat sum grouped on Agent Names
df12 = agent_perf_chats.groupby('Agent Name')['Total Chats'].sum() 
df12 = df12.to_frame()

# Adding a column to df12 which represents the Total number of feedback grouped on agent name
df12['Total Feedback'] = agent_perf_chats.groupby('Agent Name')['Total Feedback'].sum() 

# adding one more column which represents the chat feedback percentage
df12['Chat-feedback Percentage'] = (df12['Total Feedback'] / df12['Total Chats']) * 100
df12
```

## 13. Total contributation hour for each and every agents weekly basis

## 14. Total percentage of active hour for a month
